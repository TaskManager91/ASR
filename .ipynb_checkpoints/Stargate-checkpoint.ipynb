```python
import spacy
import pickle
import numpy as np

from datasets import load_from_disk, load_dataset

from tqdm import tqdm
from collections import Counter
from datetime import datetime
```

```python
import glob

from chardet.universaldetector import UniversalDetector

detector = UniversalDetector()

episode_counter = 0
file_name_counter = 0

full_text = open("Stargate/full.txt", "w", encoding="utf-8")

for season in glob.iglob("Stargate/txt/*"):
    for episode in glob.iglob(season +"/*"):
        episode_counter = episode_counter+1
        for file_name in glob.iglob(episode +"/*"):
            file_name_counter = file_name_counter +1
            
            print(file_name)
            episode_file_bytes = open(file_name, "rb")
            
            detector.reset()

            for line in episode_file_bytes:
                detector.feed(line)
                if detector.done: break
                    
            detector.close()
            print(detector.result)
            file_encoding = detector.result['encoding']
            episode_file_bytes.close()
            
            episode_file = open(file_name, "r", encoding=file_encoding)
            
            for line in episode_file:
                full_text.write(line)
                
            episode_file.close()
            
full_text.close()
print(episode_counter)
print(file_name_counter)
```

```python
import re
chars_to_keep = '[^A-Za-zäüö ]+'

full_clean = open("Stargate/full_clean.txt", "w", encoding="utf-8")

full_text = open("Stargate/full.txt", "r", encoding="utf-8")

for line in full_text:
    if line.strip():
        clean_string = re.sub(chars_to_keep, '', line).lower() + "\n"
        full_clean.write(clean_string)
    
full_clean.close()
full_text.close()
```

```python
import spacy
from tqdm import tqdm


spacy.prefer_gpu()
nlp = spacy.load("de_dep_news_trf")
nlp.max_length = 17000000

sg_trf = []

full_clean = open("Stargate/full_clean.txt", "r", encoding="utf-8")

for i, line in enumerate(tqdm(full_clean)):
    sg_trf.append(nlp(line, disable = ['ner', 'parser']))

pickle.dump(sg_trf, open("Stargate/sg_trf.p", "wb"))
```
